```python
import os
import pandas as pd
```

```python
acc_data_df = pd.read_csv('acc_players-2324F.csv', sep=',')
acc_data_df.columns = acc_data_df.iloc[0]
acc_data_df = acc_data_df.drop(0)
```

```python
print("Acc Data:")
acc_data_df.head()
```

```python
acc_data_df['PTS'] = pd.to_numeric(acc_data_df['PTS'], errors='coerce')
totalPoints = acc_data_df['PTS'].sum()
print("Total Points All Players:")
print(totalPoints)
```

```python
acc_data_df['MP'] = pd.to_numeric(acc_data_df['MP'], errors='coerce')
mostMinutes = acc_data_df.loc[acc_data_df['MP'].idxmax(), 'Player']
print("Player with the most minutes played:")
print(mostMinutes)
```

```python
acc_data_df['TRB'] = pd.to_numeric(acc_data_df['TRB'], errors='coerce')
topRebounds = acc_data_df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
print("Top 5 Rebound Leaders:")
print(topRebounds)
```

```python
minutes_df = acc_data_df[acc_data_df['MP'] > 500]
print("Players who played more than 500 minutes")
minutes_df
```

```python
highest_assits = minutes_df.loc[minutes_df['AST'].idxmax(), 'Player']
print("Player with the highest total assists:")
highest_assits

```

```python
acc_data_df['AST'] = pd.to_numeric(acc_data_df['AST'], errors='coerce')
assits_leaders = minutes_df[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)
print("Top 3 Assist Leads in the League:")
assits_leaders
```

```python
minutes_df['BLK'] = pd.to_numeric(acc_data_df['BLK'], errors='coerce')
```

```python
shot_blockers = minutes_df[['Player', 'BLK']].nlargest(3, 'BLK')
print("Top 3 Blockers in the League:")
shot_blockers
```

```python
total_points_school = acc_data_df.groupby('School')['PTS'].sum()
print("Total Points by School:")
total_points_school
```

```python
total_assists_school = acc_data_df.groupby('School')['AST'].sum()
print("Total Assists by School:")
total_assists_school
```

```python
total_points_3school = acc_data_df.groupby('School')['PTS'].sum().nlargest(3)
print("Top 3 Schools by Points Scored:")
total_points_3school
```

```python
top5 = acc_data_df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)
print("Top 5 Scorers:")
top5
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='PTS', y='Player', data=top5)
plt.title('Top 5 Scorers')
plt.show()
```

```python
correlation_fg_ws = acc_data_df['FG%'].corr(acc_data_df['WS'])
print("Correlation between Field Goals and Win Shares:")
print(correlation_fg_ws)

acc_data_df['FG%'] = pd.to_numeric(acc_data_df['FG%'], errors='coerce')
acc_data_df['WS'] = pd.to_numeric(acc_data_df['WS'], errors='coerce')


scatterplot_fg_ws = plt.figure(figsize=(10,6))
plt.scatter(acc_data_df['FG%'], acc_data_df['WS'])
plt.show()
```
